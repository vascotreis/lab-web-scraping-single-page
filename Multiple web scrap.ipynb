```python
# importing necessary libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd

# storing the link to wikipedia page of Python (programming language) in a variable
url = "https://en.wikipedia.org/wiki/Python_(programming_language)"

# sending a GET request to the link and storing the response
response = requests.get(url)

# checking the status code of the response to ensure that the request was successful
response.status_code

# parsing the contents of the response using BeautifulSoup and storing it in a variable
pythonwikisoup = BeautifulSoup(response.content, 'html.parser')

# pretty-printing the parsed content to make it easier to read
print(pythonwikisoup.prettify())

# storing the link to a US Code website in a variable
url = "http://uscode.house.gov/download/download.shtml"

# sending a GET request to the link and storing the response
response = requests.get(url)

# checking the status code of the response to ensure that the request was successful
response.status_code

# parsing the contents of the response using BeautifulSoup and storing it in a variable
uscodesoup = BeautifulSoup(response.content, 'html.parser')

# pretty-printing the parsed content to make it easier to read
print(uscodesoup.prettify())

# extracting text of all elements with class "usctitlechanged"
for i in uscodesoup.select('div.usctitlechanged'):
    print(i.get_text(strip=True))

# storing the link to the FBI's "Ten Most Wanted" page in a variable
url = "https://www.fbi.gov/wanted/topten"

# sending a GET request to the link and storing the response
response = requests.get(url)

# checking the status code of the response to ensure that the request was successful
response.status_code

# parsing the contents of the response using BeautifulSoup and storing it in a variable
mostwantedsoup = BeautifulSoup(response.content, 'html.parser')

# pretty-printing the parsed content to make it easier to read
print(mostwantedsoup.prettify())

# creating an empty list to store the names of the "Ten Most Wanted" criminals
mostwantedcriminals = []

# extracting the text from all elements with tag "h3" within a "li" tag
for i in mostwantedsoup.select('li h3'):
    mostwantedcriminals.append(i.get_text(strip=True))

# converting the list to a pandas dataframe for easier data manipulation and analysis
df = pd.DataFrame(mostwantedcriminals, columns=['Criminal Name'])
```
